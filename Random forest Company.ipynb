```python
data.loc[data.Sales <= 7.5,"Sales"]=0
data.loc[data.Sales > 7.5,"Sales"]=1
```

```python
data= pd.concat([data.drop('US', axis=1), pd.get_dummies(data['US'])], axis=1)
```

```python
data= pd.concat([data.drop('Urban', axis=1), pd.get_dummies(data['Urban'])], axis=1)
```

```python
data= pd.concat([data.drop('ShelveLoc', axis=1), pd.get_dummies(data['ShelveLoc'])], axis=1)
```

```python
data
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
x=data.drop('Sales',axis=1)
```

```python
scale=scaler.fit_transform(x)
```

```python
scale
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y = train_test_split(scale,data.Sales,test_size=0.3)
```

```python
print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
print(test_y.shape)
```

```python
train_y
```

```python
from sklearn.ensemble import RandomForestClassifier
model  =  RandomForestClassifier()
```

```python
model.fit(train_x,train_y)
```

```python
model.estimators_ 
```

```python
model.classes_
```

```python
model.n_classes_
```

```python
model.predict(train_x)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(train_y,model.predict(train_x))
```
